### 2. Read the Data and Classify it

```python
tweets = pd.read_csv('Tweets\\final_tweets_cleaned.csv')
tweets.columns
```

```python
tweets = tweets.drop(['Unnamed: 0'], axis=1)
```

```python
display(tweets.head())
display(tweets.tail())
```

```python
# Test run 
tweet = 'كيفك'
print(predict(tweet))

tweets.shape[0]
```

```python
tweets.isna().sum()
```

```python
tweets = tweets.dropna(subset=['Tweet'])
```

```python
size = 50
chunks = [tweets[i:i+size].copy() for i in range(0, tweets.shape[0], size)]
```

```python
no_chunks = len(chunks)
no_chunks
```

```python
for i in range(0, no_chunks):
    chunks[i]['Class'] = chunks[i]['Tweet'].map(predict)
    print(f'__________ lap {i+1} of {no_chunks} __________')
```

### 3. Store the Classified Data

```python
tweets = pd.concat(chunks)
```

```python
display(tweets.head())
display(tweets.tail())
```

```python
tweets = tweets.reset_index()
```

```python
display(tweets.tail())
```
